```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
df = pd.read_csv("salaries-by-college-major.csv")
print("üîç ÂâçÂπæÁ≠ÜË≥áÊñôÈ†êË¶ΩÔºö")
print(df.head())
print("\nüîß Ê∏ÖÁêÜÂâçÔºö", df.isnull().sum())
df = df.dropna()
print("‚úÖ Ê∏ÖÁêÜÂæåÔºö", df.isnull().sum())
top_majors = df.sort_values("Mid-Career Median Salary", ascending=False).head(5)
bottom_majors = df.sort_values("Mid-Career Median Salary", ascending=True).head(5)
print("\nüí∞ Ëñ™Ë≥áÊúÄÈ´òÂâç‰∫îÂêçÁßëÁ≥ªÔºö")
print(top_majors[["Undergraduate Major", "Mid-Career Median Salary"]])
print("\nü™ô Ëñ™Ë≥áÊúÄ‰ΩéÂâç‰∫îÂêçÁßëÁ≥ªÔºö")
print(bottom_majors[["Undergraduate Major", "Mid-Career Median Salary"]])
plt.figure()
sns.barplot(x="Mid-Career Median Salary", y="Undergraduate Major", data=top_majors, palette="Greens_r")
plt.title("Top 5 Majors by Mid-Career Median Salary")
plt.xlabel("Salary ($)")
plt.ylabel("Major")
plt.show()
df["Salary Growth"] = df["Mid-Career 90th Percentile Salary"] - df["Starting Median Salary"]
top_growth = df.sort_values("Salary Growth", ascending=False).head(5)
print("\nüìà Ëñ™Ë≥áÊàêÈï∑ÊΩõÂäõÂâç‰∫îÂêçÁßëÁ≥ªÔºö")
print(top_growth[["Undergraduate Major", "Starting Median Salary", "Mid-Career 90th Percentile Salary", "Salary Growth"]])
plt.figure()
sns.barplot(x="Salary Growth", y="Undergraduate Major", data=top_growth, palette="Blues_r")
plt.title("Top 5 Majors by Salary Growth Potential")
plt.xlabel("Salary Growth ($)")
plt.ylabel("Major")
plt.show()
plt.figure(figsize=(14, 6))
sns.boxplot(data=df[["Starting Median Salary", "Mid-Career Median Salary"]])
plt.title("Salary Distribution: Starting vs Mid-Career")
plt.ylabel("Salary ($)")
plt.xticks([0, 1], ["Starting", "Mid-Career"])
plt.show()
stem_keywords = ["Engineering", "Science", "Math", "Physics", "Chemistry", "Computer", "Biology"]
def classify_major(major):
    for keyword in stem_keywords:
        if keyword.lower() in major.lower():
            return "STEM"
    return "Non-STEM"
df["Major Category"] = df["Undergraduate Major"].apply(classify_major)
avg_salary_by_cat = df.groupby("Major Category")[["Starting Median Salary", "Mid-Career Median Salary"]].mean()
print("\nüìä STEM vs Èùû STEM Ëñ™Ë≥áÊØîËºÉÔºö")
print(avg_salary_by_cat)
avg_salary_by_cat.plot(kind="bar", colormap="viridis")
plt.title("STEM vs Non-STEM: Average Salaries")
plt.ylabel("Salary ($)")
plt.xticks(rotation=0)
plt.legend(loc="upper left")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("salaries_by_college_major.csv").dropna()
stem_keywords = ["Engineering", "Science", "Math", "Physics", "Chemistry", "Computer", "Biology", "Technology"]
def classify_major(major):
    for keyword in stem_keywords:
        if keyword.lower() in major.lower():
            return "STEM"
    return "Non-STEM"
df["Major Category"] = df["Undergraduate Major"].apply(classify_major)
salary_comparison = df.groupby("Major Category")[["Starting Median Salary", "Mid-Career Median Salary"]].mean().reset_index()
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))
salary_melted = salary_comparison.melt(id_vars="Major Category", var_name="Salary Type", value_name="Salary")
sns.barplot(x="Salary Type", y="Salary", hue="Major Category", data=salary_melted, palette="Set2")
plt.title("Salary Comparison: STEM vs Non-STEM Majors")
plt.ylabel("Salary ($)")
plt.xlabel("Salary Type")
plt.tight_layout()
plt.savefig("stem_vs_nonstem_salary_comparison.png", dpi=300, bbox_inches='tight')
from google.colab import files
files.download("stem_vs_nonstem_salary_comparison.png")
```
